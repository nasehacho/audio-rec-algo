```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
rec_df = pd.read_csv('../data/recommend.csv')
```

```python
rec_df.head()
```

```python
rec_df = rec_df.drop(['Unnamed: 0','artist_name','lyrics','track_name','genre', 'topic'], axis=1)
```

```python
# apply trained machine learning algorithm

kmeans = KMeans(n_clusters=3)

# save cluster assignments to new column in rec_df

rec_df['Cluster'] = kmeans.fit_predict(rec_df)
```
